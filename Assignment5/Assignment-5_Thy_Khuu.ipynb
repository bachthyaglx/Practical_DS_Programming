## Assignment 03

**Plot some histograms**

Write a function ``assignment_03`` creates a normally distributed variable with mean 5 and standard deviation 2. Plot the histogram.


```python
def assignment_03():
    mean = 5
    std_dev = 2
    
    # Generate random 10000 sample(s) for 1-dimensional arrway from univariate normal distribution
    random_data = np.random.normal(mean, std_dev, 10000)

    # Create a kernel density estimate (KDE) histogram with smoothed bins
    sns.histplot(random_data, bins=10, color='red')
    
    
assignment_03()    
```

## Assignment 04

**Plot some 2D Kernel Density Estimates**

Write a function ``assignment_04`` creates a 2D normally distributed variable with mean [0,0] and covariance matrix [[5, 2], [2, 2]]. Plot the joint kernel density estimate.


```python
def assignment_04():
    # INSERT CODE HERE
    mean = [0, 0]
    matrix = [[5, 2], [2, 2]]
    
    # Generate random sample(s) for 2D-dimensional array from multivariate normal distribution
    random_data = np.random.multivariate_normal(mean, matrix, 1000)

    # Extract x and y from the random_data
    x = random_data[:, 0] 
    y = random_data[:, 1]
    
    # Plot the joint kernel density estimate
    g = sns.jointplot(x=x, y=y, kind="kde")

    # Set labels for the plot
    plt.xlabel('x')
    plt.ylabel('y')


assignment_04()
```
